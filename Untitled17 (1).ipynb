```python
train_path = "/content/drive/MyDrive/data set/train"
val_path = "/content/drive/MyDrive/data set/val"
test_path = "/content/drive/MyDrive/data set/test"
```

```python
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_gen = ImageDataGenerator(rescale=1./255)
val_gen = ImageDataGenerator(rescale=1./255)

# Check if train_path exists
if not os.path.exists(train_path):
    print(f"Error: Training data directory not found at {train_path}. Please ensure the path is correct and the directory exists.")
    train_data = None # Or handle this error as appropriate
else:
    train_data = train_gen.flow_from_directory(
        train_path,
        target_size=(224,224),
        batch_size=8,
        class_mode='categorical'
    )

# Check if val_path exists
if not os.path.exists(val_path):
    print(f"Error: Validation data directory not found at {val_path}. Please ensure the path is correct and the directory exists.")
    val_data = None # Or handle this error as appropriate
else:
    val_data = val_gen.flow_from_directory(
        val_path,
        target_size=(224,224),
        batch_size=8
    )
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(224,224,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.GlobalAveragePooling2D(),

    layers.Dense(128,activation='relu'),
    layers.Dense(4,activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Accuracy Graph")
plt.show()
```

```python
test_gen = ImageDataGenerator(rescale=1./255)

test_data = test_gen.flow_from_directory(
    test_path,
    target_size=(224,224)
)

model.evaluate(test_data)
```
